```python
from urllib import request
import pandas as  pd
import numpy as np
```

#### <span style="color:blue"> 1. Download do Dataset do Portal de Dados Abertos </span>

```python
file_url = 'https://ons-dl-prod-opendata.s3.amazonaws.com/dataset/reservatorio/RESERVATORIOS.csv'
source_file = 'RESERVATORIOS.csv'
request.urlretrieve(file_url , source_file )
```

#### <span style="color:blue"> 2. Importar arquivo </span>

```python
dataframe = pd.read_csv(source_file, sep=';',encoding = "utf-8")

if (dataframe.empty) :
    print('DATAFRAME VAZIO')
else:
  print('shape:', dataframe.shape)
  print('columns:', dataframe.columns)
```

#### <span style="color:blue"> 5. Análise sobre os resultados - Foco em NOME DO RESERVATÓRIO</span>

<i> A função <b>value_counts()</b> do pandas é valiosíssima nas análises de dados, o que ela faz é verificar a contagem de dados iguais em um dataframe ou em uma coluna, e retorna uma Series com esse resultado.</i>  

```python
dataframe.nom_reservatorio.value_counts(ascending=False)
```

<i>A função <b>duplicated()</b> nos retorna uma série booleana denotando linhas duplicadas. Os resultados são Verdadeiro(True) e Falso(False), sendo verdadeiros os valores duplicados.</i>

```python
resultado = dataframe.duplicated(subset=['nom_reservatorio'], keep=False)
dataframe[resultado]
```
